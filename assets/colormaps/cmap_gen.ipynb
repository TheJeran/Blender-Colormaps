```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import os

# Number of colors to sample from each colormap
n_colors = 256

# Output directory
output_dir = "matplotlib"
os.makedirs(output_dir, exist_ok=True)

# Get all colormap names, excluding reversed ones
colormap_names = [name for name in plt.colormaps() if not name.endswith('_r')]

for cmap_name in colormap_names:
    cmap = cm.get_cmap(cmap_name, n_colors)
    filename = os.path.join(output_dir, f"{cmap_name}.rgb")
    
    with open(filename, 'w') as f:
        f.write(f"ncolor={n_colors}\n")
        f.write("# r g b\n")
        f.write("#Colors\n")
        for i in range(n_colors):
            r, g, b, _ = cmap(i / (n_colors - 1))  # Normalize index
            f.write(f"{r:.6f} {g:.6f} {b:.6f}\n")

print(f"✅ Generated {len(colormap_names)} .rgb files in '{output_dir}'")
```
